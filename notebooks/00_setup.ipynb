```python
from pathlib import Path

def find_root(markers=("config.json","corpus.json",".git")):
    p = Path.cwd()
    for cand in (p, *p.parents):
        if any((cand/m).exists() for m in markers):
            return cand
    return Path.cwd()

ROOT = find_root()
RAW  = ROOT/"data_raw"
CLEAN= ROOT/"data_clean"
OUT  = ROOT/"outputs"
(REPORT := ROOT/"report").mkdir(parents=True, exist_ok=True)
for d in (RAW, CLEAN, OUT):
    d.mkdir(parents=True, exist_ok=True)

print("Project root:", ROOT.resolve())
print("Folders ready:", RAW, CLEAN, OUT, REPORT, sep="\n- ")
```
